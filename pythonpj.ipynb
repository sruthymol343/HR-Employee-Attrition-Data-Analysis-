```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Sruthy\Desktop\nm internship\WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.drop_duplicates(inplace=True)
df.head()

```

```python
df['Attrition'] = df['Attrition'].replace({'Yes':1, 'No':0})
df.head()
```

```python
df.rename(columns={'MonthlyIncome':'Salary'}, inplace=True)
df.columns
```

```python
df['Experience'] = df['TotalWorkingYears']
df['SalaryLevel'] = pd.qcut(df['Salary'], 3, labels=['Low','Medium','High'])
df['ExpGroup'] = pd.cut(df['Experience'],
                       bins=[0,5,10,20,40],
                       labels=['0-5','5-10','10-20','20+'])
df.head()
```

```python
df['Attrition'] = df['Attrition'].replace({'Yes':1, 'No':0})
attrition_rate = df['Attrition'].mean() * 100
print("Attrition Rate:", attrition_rate)
```

```python
dept_attrition = df.groupby('Department')['Attrition'].mean() * 100
print(dept_attrition)
```

```python
df.groupby('ExpGroup')['Attrition'].mean() * 100
```

```python
df.groupby('JobSatisfaction')['Attrition'].mean() * 100
df.groupby('OverTime')['Attrition'].mean() * 100
```

```python
pd.crosstab(df['Department'], df['Attrition'])
```

```python
df['Department'].value_counts().plot(kind='bar')
plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
df['Attrition'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```

```python
df['Age'].plot.hist()
plt.title("Age Distribution")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='Attrition', y='Salary', data=df)
plt.title("Salary vs Attrition")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(14,10))

sns.heatmap(
    corr,
    cmap='coolwarm',
    annot=False,        # ❗ remove numbers
    linewidths=0.5      # spacing between cells
)

plt.title("Correlation Heatmap (Clean View)", fontsize=16)
plt.show()
```
